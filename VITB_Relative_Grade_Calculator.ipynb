```python
demo["total marks"] = demo["S1"] + demo["S2"] + demo["S3"] + demo["S4"] + demo["S5"]
```

```python
demo["average"] = demo["total marks"]/5
print(demo)
```

```python
# Subject columns
subjects = ["S1", "S2", "S3", "S4", "S5"]
```

```python
# Relative grading with fail condition
def get_grade(marks, mean, sd):
    # Hard fail
    if marks < 40:
        return "F"

    # Relative grading logic
    if marks >= mean + 1.5 * sd:
        return "S"
    elif marks >= mean + 0.5 * sd:
        return "A"
    elif marks >= mean:
        return "B"
    elif marks >= mean - 0.5 * sd:
        return "C"
    elif marks >= mean - 1.0 * sd:
        return "D"
    else:
        return "E"

# Calculate grade for each subject
for sub in subjects:
    mean = demo[sub].mean()
    sd = demo[sub].std()

    grade_col = f"Grade_{sub}"
    demo[grade_col] = demo[sub].apply(lambda x: get_grade(x, mean, sd))
```

```python
writer = pd.ExcelWriter('VITB_GradeSystem.xlsx')
```

```python
demo.to_excel(writer)
```
